SUBBAB SCALAR

```python
# buat variabel skalar x dan simpan suatu nilai skalar
x = 67
```

```python
# tampilkan tipe data dari x
type (x)
```

```python
# buat variabel skalar y dan simpan suatu nilai skalar
y = 76
```

```python
# tampilkan tipe data dari y
type (y)
```

```python
# jumlahkan x+y
py_sum = x + y
py_sum
```

```python
# buat variabel bernama x_float dan simpan suatu nilai desimal
x_float = 6.5
```

```python
# jumlahkan x_float dengan y
py_sum = x_float + y
py_sum
```

```python
# tampilkan tipe data dari x_float
type (x_float)
```

TENSOR FLOW

```python
# lakukan tahapan pada soal 1 dengan menggunakan tensorflow

import tensorflow as tf
```

```python
x_tf = tf.Variable(67, dtype=tf.int32)
x_tf
```

```python
x_tf.dtype
```

```python
y_tf = tf.Variable(76)
y_tf
```

```python
y_tf.dtype
```

```python
tf_sum = x_tf + y_tf
tf_sum
```

```python
x_float_tf = tf.Variable(6.5)
x_float_tf
```

```python
tf_float_sum = x_float_tf + tf.cast(y_tf, tf.float32)
tf_float_sum
```

```python
x_float_tf.dtype
```

PYTORCH

```python
# lakukan tahapan pada soal 1 dengan menggunakan pytorch

import torch
```

```python
x_pt = torch.tensor(67)
x_pt
```

```python
x_pt.dtype
```

```python
y_pt = torch.tensor(76)
y_pt
```

```python
y_pt.dtype
```

```python
pt_sum = x_pt + y_pt
pt_sum
```

```python
x_float_pt = torch.tensor(6.5)
x_float_pt
```

```python
x_float_pt.dtype
```
