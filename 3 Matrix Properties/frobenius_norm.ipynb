23. frobenius norm

Numpy

```python
# menggunakan numpy, pytoch dan tensorflow buatlah matrix ukuran 3x3 lalu hitung nilai frobenius norm
import numpy as np

X_np = np.array([[7, 6, 7], [3, 7, 6], [9, 8, 5]])
X_np
```

```python
(7**2 + 6**2 + 7**2 + 3**2 + 7**2 + 6**2 + 9**2 + 8**2 + 5**2)**(1/2)
```

```python
np.linalg.norm(X_np)
```

Pytorch

```python
import torch

X_pt = torch.tensor([[7, 6, 7], [3, 7, 6], [9, 8, 5]])
X_pt
```

```python
torch.norm(X_pt.float())
```

Tensorflow

```python
import tensorflow as tf

X_tf = tf.Variable([[7, 6, 7], [3, 7, 6], [9, 8, 5]])
X_tf
```

```python
tf.norm(tf.cast(X_tf, dtype=tf.float32))
```
