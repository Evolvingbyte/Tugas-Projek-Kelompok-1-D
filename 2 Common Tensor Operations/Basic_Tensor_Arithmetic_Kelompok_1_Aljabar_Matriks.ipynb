```python
import numpy as np
import torch
import tensorflow as tf

# CODE PADA SUB BAB SEBELUMNYA!

# Buat matrix X ukuran 2x3 menggunakan numpy
X = np.array([[1, 2, 3],
              [4, 5, 6]])

# Buat matrix X_pt ukuran 2x3 menggunakan pytorch
X_pt = torch.tensor([[1, 2, 3],
                     [4, 5, 6]])

# Buat matrix X_tf ukuran 2x3 menggunakan tensorflow
X_tf = tf.constant([[1, 2, 3],
                    [4, 5, 6]])

# Lakukan transpose pada X, X_pt, dan X_tf
X_t = X.T
X_pt_t = torch.transpose(X_pt, 0, 1)
X_tf_t = tf.transpose(X_tf)
```

```python
# Basic Tensor Arithmetic

# Lakukan perkalian matrix X, X_pt, dan X_tf dengan suatu skalar
X_kali = X * 2
X_pt_kali = X_pt * 2
X_tf_kali = X_tf * 2
```

```python
# Lakukan penjumlahan matrix X, X_pt, dan X_tf dengan suatu skalar
X_tambah = X + 2
X_pt_tambah = X_pt + 2
X_tf_tambah = X_tf + 2
```

```python
# Lakukan pengurangan matrix X, X_pt, dan X_tf dengan suatu skalar
X_kurang = X - 2
X_pt_kurang = X_pt - 2
X_tf_kurang = X_tf - 2
```

```python
# Buat matrix Y ukuran 2x3 menggunakan numpy
Y = np.array([[6, 5, 4],
              [3, 2, 1]])
```

```python
# Buat matrix Y_pt ukuran 2x3 menggunakan pytorch
Y_pt = torch.tensor([[6, 5, 4],
                     [3, 2, 1]])
```

```python
# Buat matrix Y_tf ukuran 2x3 menggunakan tensorflow
Y_tf = tf.constant([[6, 5, 4],
                    [3, 2, 1]])
```

```python
# Kalikan matrix X dengan Y
hasil_np = X * Y
```

```python
# Kalikan matrix X_pt dengan Y_pt
hasil_pt = X_pt * Y_pt
```

```python
# Kalikan matrix X_tf dengan Y_tf
hasil_tf = X_tf * Y_tf
```
