```python
# Tensor Transposition

# buat matrix X dengan ukuran 2x3 menggunakan numpy
import numpy as np
import torch
import tensorflow as tf

X = np.array([[1, 2, 3],
              [4, 5, 6]])
```

```python
# buat matrix X_pt dengan ukuran 2x3 menggunakan pytorch
X_pt = torch.tensor([[1, 2, 3],
                     [4, 5, 6]])
```

```python
# buat matrix X_tf ukuran 2x3 menggunakan tensorflow
X_tf = tf.constant([[1, 2, 3],
                    [4, 5, 6]])
```

```python
# Lakukan transpose pada X, X_pt, dan X_tf
X_t = X.T
X_pt_t = torch.transpose(X_pt, 0, 1)
X_tf_t = tf.transpose(X_tf)
```
