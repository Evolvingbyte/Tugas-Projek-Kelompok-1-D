### 1. buat matrix X dengan ukuran 2x3 menggunakan numpy

```python
import numpy as np
```

```python
print("Matriks X (Numpy) dengan ukuruan 2x3:")
X = np.array([[69, 42, 2], [1, 6, 7]])
X
```

- hitung berapa hasil penjumlahan elemen matrix X dengan fungsi sum

```python
print("hasil penjumlahan elemen matrix X dengan fungsi sum:")
X.sum()
```

- buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow

```python
import tensorflow as tf
```

```python
print("Matriks X_tf (TensorFlow) dengan ukuran 2x3:")
X_tf = tf.Variable([[69, 42, 2], [1, 6, 7]])
X_tf
```

### 2. hitung hasil penjumlahan seluruh elemen matrix X

```python
print("hasil penjumlahan seluruh elemen matrix X:")
X.sum()
```

- hitung hasil penjumlahan seluruh elemen matrix X_pt.

Disini karena tidak ada soal untuk membuat variabel matriks pytorch, maka saya tambahkan dulu

```python
import torch
```

```python
print("Matriks X_pt (PyTorch) dengan ukuran 2x3: ")
X_pt = torch.tensor([[69, 42, 2], [1, 6, 7]])
X_pt
```

```python
print("hasil penjumlahan seluruh elemen matrix X_pt:")
X_pt.sum()
```

- hitung hasil penjumlahan seluruh elemen matrix X_tf

```python
print("hasil penjumlahan seluruh elemen matrix X_tf:")
X_tf.numpy().sum()
```

### 3. hitung hasil penjumlahan matrix X berdasarkan row nya

```python
print("hasil penjumlahan matrix X:")
X.sum(axis=1)
```

- hitung hasil penjumlahan matrix X_pt berdasarkan row nya

```python
print("hasil penjumlahan matrix X_pt:")
torch.sum(X_pt, 1)
```

- hitung hasil penjumlahan matrix X_tf berdasarkan row nya

```python
print("hasil penjumlahan matrix X_tf: ")
tf.reduce_sum(X_tf, 1)
```

### 4. hitung hasil penjumlahan matrix X berdasarkan kolom nya

```python
print("hasil penjumlahan matrix X")
X.sum(axis=0)
```

- hitung hasil penjumlahan matrix X_pt berdasarkan kolom nya

```python
print("hasil penjumlahan matrix X_pt: ")
torch.sum(X_pt, 0)
```

- hitung hasil penjumlahan matrix X_tf berdasarkan kolom nya

```python
print("hasil penjumlahan matrix X_tf: ")
tf.reduce_sum(X_tf, 0)
```

### 5. hitung nilai minimal dari matrix X, X_pt dan X_tf

```python
print("Nilai minimal dari X:")
X.min()
```

```python
print("Nilai minimal dari X_pt:")
X_pt.min()
```

```python
print("Nilai minimal dari X_tf:")
tf.reduce_min(X_tf)
```

- hitung nilai maksimal dari matrix X, X_pt dan X_tf

```python
print("Nilai maksimal dari X:")
X.max()
```

```python
print("Nilai maksimal dari X_pt:")
X_pt.max()
```

```python
print("Nilai maksimal dari X_tf:")
tf.reduce_max(X_tf)
```

- hitung nilai rata-rata dari matrix X, X_pt dan X_tf

```python
print("Nilai rata-rata dari X:")
X.mean()
```

```python
print("Nilai rata-rata dari X_pt:")
X_pt.float().mean()
```

```python
print("Nilai rata-rata dari X_tf):")
tf.reduce_mean(tf.cast(X_tf, tf.float32))
```

### 6. hitung hasil perkalian seluruh elemen matrix X

```python
print("hasil perkalian seluruh elemen matrix X:")
X.prod()
```

- hitung hasil perkalian seluruh elemen matrix X_pt

```python
print("hasil perkalian seluruh elemen matrix X_pt:")
X_pt.prod()
```

- hitung hasil perkalian seluruh elemen matrix X_tf

```python
print("hasil perkalian seluruh elemen matrix X_tf:")
tf.reduce_prod(X_tf)
```
