### 1. buat visualisasi dari persamaan berikut:

### y = 2x + 1 dan 3x - y = 5

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000)
```

```python
# Persamaan 1
y1 = 2*x + 1
```

```python
# Persamaan 2
y2 = 3*x - 5
```

```python
fig, ax = plt.subplots()
ax.set_xlim([0, 7.5])
ax.set_ylim([0, 17])

plt.xlabel('x')
plt.ylabel('y')

ax.plot(x, y1, c='green')
ax.plot(x, y2, c='brown')

intersection_x = 6
intersection_y = 13

plt.axvline(x=intersection_x, color='purple', linestyle='--')
plt.axhline(y=intersection_y, color='purple', linestyle='--')

plt.show()
```
