```python
import numpy as np
import torch as tr
import tensorflow as tf
```

Soal 1

```python
# buat vector x dan y dengan ukuran 1x3 menggunakan numpy
x = np.array([1,2,3])
y = np.array([4,5,6])
print (f"vector x adalah : {x}")
print (f"vector y adalah : {y}")
```

```python
np.dot(x,y)
```

```python
# buat vector x_pt dan y_pt dengan ukuran 1x3 menggunakan pytorch
x_pt = tr.tensor([1,2,3], dtype=tr.float32)
y_pt = tr.tensor([4,5,6], dtype=tr.float32)
print (f"vector x_pt adalah : {x_pt}")
print (f"vector y_pt adalah : {y_pt}")
```

```python
# buat vector x_tf dan y_tf dengan ukuran 1x3 menggunakan tensorflow
x_tf = tf.constant([1,2,3], dtype=tf.float32)
y_tf = tf.constant([4,5,6], dtype=tf.float32)
print (f"vector x_tf adalah : {x_tf}")
print (f"vector y_tf adalah : {y_tf}")
```

Soal 2

```python
# hitung perkalian dot product antara vector x dan vector y
print("Dot Product NumPy :", np.dot(x, y))
```

```python
# hitung perkalian dot product antara vector x_pt dan vector y_pt
print("Dot Product Pytorch:" , tr.dot(x_pt, y_pt).item())
```

```python
# hitung perkalian dot product antara vector x_tf dan vector y_tf
print("Dot Product TensorFlow :", tf.tensordot(x_tf, y_tf, axes=1).numpy())
```
